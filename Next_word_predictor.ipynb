# Import Libraries

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index               # check numbers assign to words
```

```python
for sentences in faqs.split('\n'):        # split the dataset to sentences on basis of next line    
    print(sentences)
```

# Assign numbers to words

```python
# now on basis of old numbers,sentences are converting to numbers
input_sequence = []
for sentences in faqs.split('\n'):                                           # split the dataset to sentences on basis of next line 
    
    tokenized_sentence=tokenizer.texts_to_sequences([sentences])[0]          # to avoid double list [[93,1,13]], we use [0]
    #print(tokenized_sentence)

    for i in range(1, len(tokenized_sentence)):
            input_sequence.append(tokenized_sentence[:i+1])                #Example:  tokenized_sentence = ["I", "like", "to", "learn", "Python"].

                                                                                #    Iteration 1 (i=1): tokenized_sentence[:2] → ["I", "like"]

                                                                                 #   Iteration 2 (i=2): tokenized_sentence[:3] → ["I", "like", "to"]

                                                                                 #   Iteration 3 (i=3): tokenized_sentence[:4] → ["I", "like", "to", "learn"]

                                                                                  #  Iteration 4 (i=4): tokenized_sentence[:5] → ["I", "like", "to", "learn", "Python"]'''
            
```

```python
input_sequence
```

```python
# now we will find the senetnce of maximum langth  bcz we need to set equal size of all inputs.
# find max length and zdd 0 padding to other all sentences to make size equal of all senetnces

#[len(x) for x in input_sequence]                         #  [len(x) for x in input_sequence check lenths of sentences
max_len = max([len(x) for x in input_sequence])        # We find the Max length of all  sequences
max_len
```

```python
# Add  Zeros_padding by keras pad_sequences function
from  tensorflow.keras.preprocessing.sequence import pad_sequences
# pad_sequences has 3 parameters 1-input ,2= max-padding-length , 3 -pre-padding/post-paddind
padded_input_sequences = pad_sequences(input_sequence , maxlen = max_len , padding = 'pre')
padded_input_sequences
```

```python
# separate the Input and Output   
# E.g. [  0,   0,   0, ...,   0,  93,   1], input=[  0,   0,   0, ...,   0,  93], output=[  1],
# E.g. [  0,   0,   0, ..., 279,  18, 280], input=[  0,   0,   0, ..., 279,  18], output=[ 280],

X = padded_input_sequences[: , : -1]
X
```

```python
Y = padded_input_sequences[: , -1]
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
# total number of words
vocab_size= len(tokenizer.word_index)
vocab_size
```

```python
# make the one-hot-Encoding of Y,,because Y is scaller 
from tensorflow.keras.utils import to_categorical
Y = to_categorical(Y , num_classes = 283)                 # words_total=282,we add 283 because the Hot-vector starts from the 0
```

```python
Y.shape
Y
```

```python
# import layers

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM,  Dense
```

# Model 

```python
#   Model Define
input_length= 56
embedding_dim = 100
vocab_size = 283

model= Sequential()
#model.add(Embedding(input_dim = vocab_size, output_dim = embedding_dim, input_length=56)) 
model.add(Embedding(283, 100, input_length=56)) 

# Embedding(Total_vocabulary_words, embedding_shape/output_dense_vector , total_words_in_every_row)

model.add(LSTM(150,  return_sequences=False))                                              # 150 nodes in LSTM_gate
model.add(Dense(vocab_size, activation = 'softmax'))                         #Dense(neurons must equal to total words in Vocavularly)
```

# Build Model

```python
 # Build the Model
model.build(input_shape=(None, input_length))  # Define batch size and input length explicitly
                                               # None means the batch size can vary, 56 features per input
```

# Compile Model

```python
#  Compile model

model.compile(loss='categorical_crossentropy', optimizer= 'adam', metrics = ['accuracy'])
```

# Model Summary

```python
model.summary()
```

```python
# Model Fitting
model.fit(X, Y, epochs=100)
```

# Prediction of Model

```python
# Test

text = "mail"
#   3-steps for word  -Tokenizer -padding  -predict

# Tokenizer
token_text = tokenizer.texts_to_sequences([text])[0]

# padding
padded_token_text = pad_sequences([token_text], maxlen= 56, padding='pre')

# predict             send padded_token_text to the model
model.predict(padded_token_text)                         # it returns the all words probablity

```

```python
#  now from all words probablities, we choose Highest

import numpy as np
pos = np.argmax(model.predict(padded_token_text))
print(pos)
```

```python
for word, index in tokenizer.word_index.items():
 if index == pos:
    print(word)
```
